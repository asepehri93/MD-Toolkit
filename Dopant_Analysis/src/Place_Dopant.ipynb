```python
class Place_Dopants():

    def __init__(self, method, file_name, orig_atom, dopant, percentage):
        self.method = method
        self.file_name = file_name
        self.orig_atom = orig_atom
        self.dopant = dopant
        self.num_orig = 0
        self.orig_atom_coords = []
        self.percentage = percentage




    def file_handler(self):
        if not os.path.exists('dopand_analysis'):
            os.mkdir('dopant_analysis')

        new_fname = 'dopant_analysis/' + self.file_name.split('.')[0] + '.txt'
        shutil.copy(self.file_name, new_fname)
        
        with open(new_fname, 'r') as f:
            self.lines = f.readlines()
        
        self.num_atoms = int(self.lines[0].strip())
        
        for i, line in enumerate(self.lines):
            if line.startswith(self.orig_atom):
                self.num_orig += 1
                self.orig_atom_coords.append((int(i), float(line.split()[1]),
                                              float(line.split()[2]),
                                              float(line.split()[3])))
        
        self.count = int(self.percentage * self.num_orig / 100)
        self.orig_atom_coords = np.array(self.orig_atom_coords, dtype=object)




    def place_random(self):
        new_file = ''
        sample_list = random.sample(self.orig_atom_coords[:,0], k=self.count)

        for i, line in enumerate(self.lines):
            if i in sample_list:
                line = line.replace(self.orig_atom, self.dopant)
                new_file += line
            else:
                new_file += line

        with open('dopant_analysis/' + str(self.percentage) + '_random.txt', 'a') as f:
            f.write(new_file)




    def place_clustered(self):
        new_file = ''
        selected_atoms = []
        initial_atom_index = random.sample(range(len(self.orig_atom_coords)), k=1)       #index WRT self.orig_atom_coords
        selected_atoms.append(initial_atom_index[0])

        for c in range(self.count-1):
            remaining_arr = np.delete(self.orig_atom_coords, selected_atoms, axis=0)
            distances_arr = np.zeros((remaining_arr.shape[0],len(selected_atoms)))
            for i, atom in enumerate(selected_atoms):
                atom_coords = self.orig_atom_coords[atom, 1:]
                distances_arr[:,i] = (np.sum((remaining_arr[:, 1:] - atom_coords)**2
                                        , axis=1) ** 0.5)
            
            new_atom_index = np.sum(distances_arr, axis=1).argmin()          #index WRT remaining_arr
            idx = np.where(self.orig_atom_coords[:,0] == remaining_arr[new_atom_index, 0])     #index WRT self.orig_atom_coords
            selected_atoms.append(int(idx[0]))

        real_indices = []
        for atom in selected_atoms:
            real_indices.append(self.orig_atom_coords[atom, 0])           #index WRT lines
        
        
        for i, line in enumerate(self.lines):
            if i in selected_atoms:
                line = line.replace(self.orig_atom, self.dopant)
                new_file += line
            else:
                new_file += line

        with open('dopant_analysis/' + str(self.percentage) + '_clustered.txt', 'a') as f:
            f.write(new_file)




    def place_dispersed(self):
        new_file = ''
        selected_atoms = []
        initial_atom_index = random.sample(range(len(self.orig_atom_coords)), k=1)
        selected_atoms.append(initial_atom_index[0])

        for c in range(self.count-1):
            remaining_arr = np.delete(self.orig_atom_coords, selected_atoms, axis=0)
            distances_arr = np.zeros(remaining_arr.shape[0],len(selected_atoms))
            for i, atom in enumerate(selected_atoms):
                atom_coords = self.orig_atom_coords[atom, 1:]
                distances_arr[:,i] = (np.sum((remaining_arr[:, 1:] - atom_coords)**2
                                        , axis=1) ** 0.5)
            
            new_atom_index = distances_arr.min(axis=1).argmax()
            selected_atoms.append(remaining_arr[new_atom_index, 1])

        for i, line in enumerate(self.lines):
            if i in selected_atoms:
                line = line.replace(self.orig_atom, self.dopant)
                self.new_file += line
            else:
                self.new_file += line

        with open('dopant_analysis/' + str(self.percentage) + '_dispersed.txt', 'a') as new_file:
            new_file.write(new_file)
```
